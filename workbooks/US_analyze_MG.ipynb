```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
import json
```

```python
# reading the csv
us_df = pd.read_csv("../resources/MG_USvideos_MG.csv", encoding="ISO-8859-1")
us_df.head(10)
```

```python
# reading the json file
file = open("../resources/US_category_id_MG.json")
us_json = json.load(file)
```

```python
# looping through the json file and replacing the category id with the category name
length = len(us_json["items"])
print(length)

us_df["category_id"] = us_df["category_id"].astype(str)

for x in range(length):
    id_number = us_json["items"][x]["id"]
    category_name = us_json["items"][x]["snippet"]["title"]
    us_df["category_id"] = us_df["category_id"].replace({f"{id_number}": f"{category_name}"})
    
us_df["category_id"] = us_df["category_id"].replace({"29": "Nonprofits & Activism"})
    
us_df.head()
```

```python
# this shows how many videos of each category is present in the data
us_count_df = pd.DataFrame(us_df.groupby("category_id").count(), columns=["video_id"])
us_count_df = us_count_df.rename(columns={"video_id": "count"})
us_count_df.sort_values(by=["count"], ascending=False)
```

```python
# separate dataframe using just the numerical value columns
# this is probably not too needed
us_dateview_df = pd.DataFrame(us_df, columns=["trending_date", "views", "likes", "dislikes", "comment_count"])
```

```python
# views vs likes on all videos
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(us_dateview_df["views"], us_dateview_df["likes"])
regress_value = us_dateview_df["views"] * slope + intercept
line_eq = f"y = {round(slope, 4)}x + {round(intercept, 0)}"
plt.plot(us_dateview_df["views"], regress_value, "r-")
plt.annotate(line_eq, (1e7, 3e6), fontsize=14, color="red")
plt.scatter(us_dateview_df["views"], us_dateview_df["likes"])
plt.xlabel("views")
plt.ylabel("likes")
plt.show()
```

```python
# views vs dislikes on all videos
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(us_dateview_df["views"], us_dateview_df["dislikes"])
regress_value = us_dateview_df["views"] * slope + intercept
line_eq = f"y = {round(slope, 4)}x + {round(intercept, 0)}"
plt.plot(us_dateview_df["views"], regress_value, "r-")
plt.annotate(line_eq, (1e7, 4e5), fontsize=14, color="red")
plt.scatter(us_dateview_df["views"], us_dateview_df["dislikes"])
plt.xlabel("views")
plt.ylabel("dislikes")
plt.show()
```

```python
# likes vs dislikes on all videos
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(us_dateview_df["likes"], us_dateview_df["dislikes"])
regress_value = us_dateview_df["likes"] * slope + intercept
line_eq = f"y = {round(slope, 4)}x + {round(intercept, 0)}"
plt.plot(us_dateview_df["likes"], regress_value, "r-")
plt.annotate(line_eq, (1e6, 4e5), fontsize=14, color="red")
plt.scatter(us_dateview_df["likes"], us_dateview_df["dislikes"])
plt.xlabel("likes")
plt.ylabel("dislikes")
plt.show()
```

```python
# shows which channel
us_channel_df = pd.DataFrame(us_df.groupby("channel_title").count(), columns=["video_id"])
us_channel_df = us_channel_df .rename(columns={"video_id": "count"})
us_channel_df.sort_values(by=["count"], ascending=False).head(10)

```

```python
us_channel_df = us_df.groupby(us_df['channel_title']).agg({'likes': ['sum'],
                                                           'dislikes': ['sum'],
                                                           'views': ['sum']})
us_channel_df.columns = ['likes', 'dislikes', 'views']

us_channel_df.sort_values(by=["dislikes"], ascending=False).head(10)
```

```python
likes = us_channel_df['likes']
dislikes =  us_channel_df['dislikes']
views = us_channel_df['views']
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(likes, dislikes, views, 
           linewidths=1, alpha=.8,
           edgecolor='k',
           s = 200,
           c= views)
plt.show()
```

```python
plt.scatter(views, dislikes, marker="o")
# plt.ylim(35,46)
plt.xlabel("views")
plt.ylabel("dislikes")
plt.show()
```

```python
diff = us_channel_df['likes'] - us_channel_df['dislikes']
views = us_channel_df['views']
plt.scatter(views, diff, marker="x")
# plt.ylim(35,46)
plt.xlabel("views")
plt.ylabel("likes-dislikes")
plt.show()
```

```python
# Top5 
us_channel_df['likes-dislikes'] = diff
top_5 = us_channel_df.sort_values(by=["likes-dislikes"], ascending=False).head(5)
top_5
```

```python
# remover outliers
us_channel_clean_df = us_channel_df.loc[(us_channel_df.index != "ibighit") & (us_channel_df.index != "ChildishGambinoVEVO") ]
us_channel_clean_df.sort_values(by=["likes-dislikes"], ascending=False).head(5)
```

```python
diff = us_channel_clean_df['likes-dislikes']
views = us_channel_clean_df['views']
plt.scatter(views, diff, marker="x")
# plt.ylim(35,46)
plt.xlabel("views")
plt.ylabel("likes-dislikes")
plt.show()
```

```python
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(us_channel_clean_df['views'], us_channel_clean_df['likes-dislikes'])
regress_value = us_channel_clean_df['views'] * slope + intercept
line_eq = f"y = {round(slope, 4)}x + {round(intercept, 0)}"
plt.plot(us_channel_clean_df['views'], regress_value, "r-")
plt.annotate(line_eq, (1e7, 4e5), fontsize=14, color="red")
plt.scatter(us_channel_clean_df['views'], us_channel_clean_df['likes-dislikes'])
plt.xlabel("views")
plt.ylabel("dislikes")
plt.show()
```
