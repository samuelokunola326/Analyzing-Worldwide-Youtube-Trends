```python
country = ["Canada", "Denmark", "France", "G.B.", "India", "Japan", "Korea", "Mexico", "Russia", "U.S."]
views = []
reactions = []

reactions.append(ca["views"].sum())
reactions.append(de["views"].sum())
reactions.append(fr["views"].sum())
reactions.append(gb["views"].sum())
reactions.append(ind["views"].sum())
reactions.append(jp["views"].sum())
reactions.append(kr["views"].sum())
reactions.append(mx["views"].sum())
reactions.append(ru["views"].sum())
reactions.append(us["views"].sum())

views.append(sum((ca["likes"] + ca["dislikes"] + ca["comment_count"]).tolist()))
views.append(sum((de["likes"] + de["dislikes"] + de["comment_count"]).tolist()))
views.append(sum((fr["likes"] + fr["dislikes"] + fr["comment_count"]).tolist()))
views.append(sum((gb["likes"] + gb["dislikes"] + gb["comment_count"]).tolist()))
views.append(sum((ind["likes"] + ind["dislikes"] + ind["comment_count"]).tolist()))
views.append(sum((jp["likes"] + jp["dislikes"] + jp["comment_count"]).tolist()))
views.append(sum((kr["likes"] + kr["dislikes"] + kr["comment_count"]).tolist()))
views.append(sum((mx["likes"] + mx["dislikes"] + mx["comment_count"]).tolist()))
views.append(sum((ru["likes"] + ru["dislikes"] + ru["comment_count"]).tolist()))
views.append(sum((us["likes"] + us["dislikes"] + us["comment_count"]).tolist()))

viewreact_dict = {"Country": country, "Views": views, "Reactions": reactions}
viewreact = pd.DataFrame(viewreact_dict)

viewreact["V-R-Ratio"] = viewreact["Views"] / viewreact["Reactions"] * 100

viewreact
```

```python
viewreact = viewreact.sort_values(by=["V-R-Ratio"], ascending=False)
explode = [0.2,0.15,0.1,0.05,0,0,0,0,0,0]

plt.pie(viewreact["V-R-Ratio"], labels=viewreact["Country"], autopct="%.1f%%",
       startangle=90, counterclock=False, explode=explode, radius=2,
        wedgeprops=dict(width=0.6, edgecolor="w"))
plt.title("Views to Reactions Ratio for each Country", y=1.5, fontsize=16)
plt.show()
```
